```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
corpus=[]
stemmer=PorterStemmer()
for i in range(0,len(messages)):
    paragraph=re.sub('[^a-zA-Z]',' ',messages['title'][i])
    paragraph=paragraph.lower()
    paragraph=paragraph.split()
    paragraph=[stemmer.stem(word) for word in paragraph if not word in stopwords.words("english")]
    paragraph= ' '.join(paragraph)
    corpus.append(paragraph)
    
corpus
```

```python
cv=TfidfVectorizer(max_features=5000,ngram_range=(1,3))
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=0)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    See full source and example: 
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# MultinomialNB with HyperParameter

```python
previous_score=0
classifier=MultinomialNB(alpha=0.1)
for alpha in np.arange(0,1,0.1):
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(X_train,y_train)
    predix=sub_classifier.predict(X_test)
    score=metrics.accuracy_score(y_test,predix)
    if score>previous_score:
        classifier=sub_classifier
        
    print(f'Alpha is {alpha} and Score is : {score}')
```
